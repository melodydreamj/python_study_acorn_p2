```python
#참조 : https://docs.google.com/document/d/16DanxYQGTlH0oOsTRte8nm8YWjjOQycgWosSGAbzixo/edit# 

import pandas as pd

baseball_player = pd.DataFrame({
    "player_Name":["권혁","이범호","필","박병호","양준혁"],
    "team_No":[1993,2001,2001,2010,None]
}
)

baseball_team = pd.DataFrame({
    "team_Name":["한화","두산","기아","넥센"],
    "team_home":["대전","서울","광주","고척"],
    "team_id":[1993,1999,2001,2010]
})
```

```python
baseball_team
```

```python
baseball_player
```

```python
#오른쪽 기준으로 함ㅇㅇ
pd.merge(baseball_player,baseball_team,how='left',left_on="team_No",right_on="team_id")#how->join방법 /left_on-> 왼쪽 기준칼럼/ right_on->오른쪽 기준칼람
```

```python
#inner join #양쪽다 공통되게 있는것만 추림
pd.merge(baseball_player,baseball_team,how='inner',left_on="team_No",right_on="team_id")
```

```python
#right outter
pd.merge(baseball_player,baseball_team,how='right',left_on="team_No",right_on="team_id")
```

```python
#left outer join + 조건
temp = pd.merge(baseball_player,baseball_team,how='left',
                left_on="team_No",right_on="team_id")#how->join방법 /left_on-> 왼쪽 기준칼럼/ right_on->오른쪽 기준칼람
temp[temp["team_No"].isnull()] 

```

```python
#right outer join + 조건
temp = pd.merge(baseball_player,baseball_team,how='right',
                left_on="team_No",right_on="team_id")#how->join방법 /left_on-> 왼쪽 기준칼럼/ right_on->오른쪽 기준칼
temp[temp["player_Name"].isnull()] 
```

```python
#union
pd.merge(baseball_player,baseball_team,how='outer',
                left_on="team_No",right_on="team_id")#how->join방법 /left_on-> 왼쪽 기준칼럼/ right_on->오른쪽 기준칼람
```

```python
temp = pd.merge(baseball_player,baseball_team,how='outer',
                left_on="team_No",right_on="team_id")#how->join방법 /left_on-> 왼쪽 기준칼럼/ right_on->오른쪽 기준칼람

temp[temp["player_Name"].isnull() | temp["team_No"].isnull()] 
```

```python
baseball_player = pd.DataFrame({
    "player_Name":["권혁","이범호","필","박병호","양준혁"],
    "team_No":[1993,2001,2001,2010,None]
},
    index = ["p_01","p_02","p_03","p_04","p_05"]
)
baseball_player.index.name = "player_id"

baseball_team = pd.DataFrame({
    "team_Name":["한화","두산","기아","넥센"],
    "team_home":["대전","서울","광주","고척"]
    #"team_id":[1993,1999,2001,2010]
},
    index = [1993,1999,2001,2010]

)
```

```python
baseball_player
```

```python
baseball_team
```

```python
#Left Outter Join
pd.merge(baseball_player,baseball_team,how='left',
         left_on="team_No",right_index=True )#how->join방법 /left_on-> 왼쪽 기준칼럼/ right_on->오른쪽 기준칼람
```

```python
#right outter
pd.merge(baseball_player,baseball_team,how='right',left_on="team_No",right_index=True)
```

```python
#inner join #양쪽다 공통되게 있는것만 추림
pd.merge(baseball_player,baseball_team,how='inner',left_on="team_No",right_index=True)
```

```python
#left outer join + 조건
temp = pd.merge(baseball_player,baseball_team,how='left',
                left_on="team_No",right_index=True)#how->join방법 /left_on-> 왼쪽 기준칼럼/ right_on->오른쪽 기준칼람
temp[temp["team_No"].isnull()] 
```

```python
#right outer join + 조건
temp = pd.merge(baseball_player,baseball_team,how='right',
                left_on="team_No",right_index=True)#how->join방법 /left_on-> 왼쪽 기준칼럼/ right_on->오른쪽 기준칼
temp[temp["player_Name"].isnull()] 
```

```python
#union
pd.merge(baseball_player,baseball_team,how='outer',
                left_on="team_No",right_index=True)#how->join방법 /left_on-> 왼쪽 기준칼럼/ right_on->오른쪽 기준칼람
```
